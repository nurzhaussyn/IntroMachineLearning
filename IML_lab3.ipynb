```python
from datetime import datetime
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data1 = pd.read_csv('operations.csv')
print(data1[:5])
```

```python
data2 = pd.read_csv('dataset.csv')
print(data2[:5])
```

```python
data3 = pd.read_csv('types.csv',sep=';')
print(data3[:5])
```

```python
data4 = pd.read_csv('codes.csv',sep=';')
print(data4[:5])
```

```python
new_data1 = pd.merge(data1, data2, on='client_id')
print(new_data1[:5])
```

```python
new_data2 = pd.merge(new_data1, data3, on='type')
new_data2[:5]
```

```python
new_data3 = pd.merge(new_data2, data4, on='code')
new_data3
```

```python
new_data3['client_id'].hist(bins=60, color = 'green')
plt.show()
# Клиенты с id:70 совершали больше операции
```

```python
new_data3['code'].hist(bins=50,color = 'blue')
plt.show()
# Операции с кодом 6000 совершались часто чем остальные
```

```python
new_data3['type'].hist(bins=50,color = 'red')
plt.show()
# Операции с 1000 часто встречаются с остальные типы
```

```python
new_data3['sum'].hist(bins=50,color = 'yellow')
plt.xlim(-1000000,1000000)
plt.show()
# Люди больше снимали свои деньги, чем пополняли 
```

```python
new_data3['status'].hist(bins=50,color = 'teal')
plt.show()
# 0 - не вернул долг
# 1 - вернул долг
# Тогда людей, которые не закрывшие свои долги, чуть больше чем закрывшие
```
